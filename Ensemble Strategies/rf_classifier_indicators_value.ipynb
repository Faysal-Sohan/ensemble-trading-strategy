```python
import pandas as pd
import numpy as np
import yfinance as yf
import pandas_ta as ta
```

```python
df = pd.read_csv('../data/S & P 50/AAPL.csv')
df.head()
```

```python
df=df[df['volume']!=0]
df.reset_index(drop=True, inplace=True)
df.isna().sum()
df.tail()
```

```python
df["RSI"] = ta.rsi(df.close, length=16)
df["CCI"] = ta.cci(df.high, df.low, df.close, length=16)
df["AO"] = ta.ao(df.high, df.low)
df["MOM"] = ta.mom(df.close, length=16)
a = ta.macd(df.close)
df = df.join(a)
df["ATR"] = ta.atr(df.high, df.low, df.close, length=16)
df["BOP"] = ta.bop(df.open, df.high, df.low, df.close, length=16)
df["RVI"] = ta.rvi(df.close)
a = ta.dm(df.high, df.low, length=16)
df = df.join(a)
a = ta.stoch(df.high, df.low, df.close)
df = df.join(a)
a = ta.stochrsi(df.close, length=16)
df = df.join(a)
df["WPR"] = ta.willr(df.high, df.low, df.close, length=16)
```

```python
df.tail(10)
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['price_after_3d'] = df.close.shift(-3)
df['price_diff_3d'] = df['price_after_3d'] - df.close
df['target'] = np.where(df['price_diff_3d'] > 1.5, 1, 0)
df['target'] = np.where(df['price_diff_3d'] < -0.5, -1, df['target'])
df.head(10)
```

```python
df.target.value_counts()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
X = df.drop(['date','price_after_3d','price_diff_3d','target'],axis=1)
X.head()
```

```python
y = df.target
y.head()
```

```python
print(X.shape, y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled[:3]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled,y, stratify=y, random_state=10)
```

```python
X_test.shape[0]/X_train.shape[0]
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)

predictions = rf_classifier.predict(X_test)

accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
```
