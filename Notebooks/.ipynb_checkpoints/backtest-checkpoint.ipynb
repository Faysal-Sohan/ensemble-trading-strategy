```python
start = '2018-01-01'
end = '2023-01-01'
capital_base = 10000
calendar = get_calendar('NYSE')
symb = "AAPL"
bundle_data = bundles.load("sp50")
window = 90

def initialize(context):
    context.i = 0
    sids = bundle_data.asset_finder.sids
    context.assets = bundle_data.asset_finder.retrieve_all(sids)


def handle_data(context, data):
    # Skip first 90 days to get full windows
    context.i += 1
    if context.i < window+1:
        return
    
    for asset in context.assets:
        df = data.history(asset,['open','high','low','close'],bar_count=window,frequency='1d')
        curr_price = data.current(asset,'price')
        result_df = ind.calc_rsi(df)
        signal = result_df.rsi_signal[-1] 

        if signal == 1:
            order_target_percent(asset, 0,stop_price=(curr_price * 0.9))
        elif signal == -1:
            order_target_percent(asset, 0.1)
        else:
            pass
    

result_rsi = run_algorithm(start=pd.Timestamp(start),
                        end=pd.Timestamp(end),
                        initialize=initialize,
                        handle_data=handle_data,
                        capital_base=capital_base,
                        trading_calendar=calendar,
                        bundle='sp50',
                        data_frequency='daily',
                        )
```

```python
start = result_rsi.index[0]
end = result_rsi.index[-1]
benchmark = yf.download('^GSPC', start=start, end=end)['Adj Close'].pct_change()
result_rsi.index = pd.to_datetime(result_rsi.index).tz_convert(None)
result_rsi.index = benchmark.index    
qs.reports.full(result_rsi['returns'], benchmark = benchmark, match_dates=True, figsize=(8, 4), df = result_rsi)
```
