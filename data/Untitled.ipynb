```python
import sys
import os
root_dir = os.getcwd().replace('\data','')
sys.path.append('../modules/')
print(root_dir)
sys.path.insert(1, root_dir)
import pandas as pd
import numpy as np
import ensemble_trading_strategy as ets
```

```python
stock_tickers = [
    "AAPL", "ABT", "ADBE", "AEP", "AES", "AFL", "ALGN", "AMD", "AMT", "AOS",
    "APA", "APD", "ARE", "AWK", "AXON", "BA", "BAC", "BAX", "BDX", "BG",
    "BKR", "BRK-B", "BX", "BXP", "CAT", "CCI", "CDNS", "CE", "CEG", "COF",
    "CPB", "CPT", "DIS", "EA", "FMC", "FOX", "GOOGL", "IFF", "KLAC", "LNT",
    "MMM", "MO", "NVDA", "STZ", "TECH", "TRGP", "VLO", "WBD", "WMB", "WMT"]
```

```python
combined_X = pd.DataFrame()
combined_y = pd.Series()
for ticker in stock_tickers:
    stock_df = pd.read_csv(f'../data/S & P 50/{ticker}.csv')
    stock_df = ets.add_features(stock_df)
    y = ets.get_target_signal_rets(stock_df)
    # stock_df['Ticker'] = ticker  # Add a column to identify the stock
    combined_X = pd.concat([combined_X, stock_df])
    combined_y = pd.concat([combined_y, y])
```
