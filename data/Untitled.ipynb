```python
import pandas as pd
import numpy as np
```

```python
stock_tickers = [
    "AAPL", "ABT", "ADBE", "AEP", "AES", "AFL", "ALGN", "AMD", "AMT", "AOS",
    "APA", "APD", "ARE", "AWK", "AXON", "BA", "BAC", "BAX", "BDX", "BG",
    "BKR", "BRK-B", "BX", "BXP", "CAT", "CCI", "CDNS", "CE", "CEG", "COF",
    "CPB", "CPT", "DIS", "EA", "FMC", "FOX", "GOOGL", "IFF", "KLAC", "LNT",
    "MMM", "MO", "NVDA", "STZ", "TECH", "TRGP", "VLO", "WBD", "WMB", "WMT"]
```

```python
combined_df = pd.DataFrame()
for ticker in stock_tickers:
    stock_df = pd.read_csv(f'../data/S & P 50/{ticker}.csv')
    stock_df['Ticker'] = ticker  # Add a column to identify the stock
    combined_df = pd.concat([combined_df, stock_df])
```

```python
combined_df.tail(10)
```

```python
combined_df.reset_index(drop=True, inplace=True)
```

```python
combined_df.to_csv('../data/dataset/combined_stocks_data.csv',index=False)
```
