```python
try:
    df = pd.read_csv('../data/S & P 500 Assets/daily/AAPL.csv')
except FileNotFoundError:
    print("File Not Fount")
df.head()
```

```python
cols = {"Date":"date","Open":"open","High":"high", "Low":"low","Close":"close","Volume":"volume"}
df.rename(columns=cols, inplace=True)
df.drop('Adj Close',axis=1,inplace=True)
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
df = df.reindex(columns=['date','open','high','low','close', 'volume'])
df['date'] = df['date'].dt.strftime('%Y-%m-%d %H:%M:%S.%f')
df.head()
```

```python
start = "2015-01-01"
end = "2017-01-01"

filtered_df = df[(df['date'] >= start) & (df['date']<=end)]
filtered_df.head()
```

```python
new_df = ind.calc_rsi(filtered_df)
new_df.tail()
```
