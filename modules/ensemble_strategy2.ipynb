```python
import numpy as np
import pandas as pd 
import indicators as ind
%load_ext autoreload
%autoreload 2
```

```python
def group3_ensemble_model_signals(df:pd.DataFrame):
    df = ind.EMA(df)
    df = ind.macd_signals(df)
    df = ind.adx_signal(df)
    df = ind.rsi_signals(df)
    df = ind.bb_signals(df)
    df = ind.keltner_channels(df)
    df = ind.stochastic_oscillator_strategy(df)
    
    return df
```

```python
signals = group3_ensemble_model_signals(df)
```

```python
signals.isna().sum()
```

```python
signals.dropna(inplace=True)
```

```python
signals['price_after_3d'] = signals.close.shift(-3)
signals['price_after_7d'] = signals.close.shift(-7)
signals.head(7)
```

```python
signals['target_3d'] = signals['price_after_3d'] - signals.close
signals['target_7d'] = signals['price_after_7d'] - signals.close
signals.head(10)
```

```python
signals['target_signal_3d'] = np.where(signals['target_3d'] > 1.5, 1, 0)
signals['target_signal_3d'] = np.where(signals['target_3d'] < -1.5, -1, signals['target_signal_3d'])

signals['target_signal_7d'] = np.where(signals['target_7d'] > 2, 1, 0)
signals['target_signal_7d'] = np.where(signals['target_7d'] < -2, -1, signals['target_signal_7d'])
```
